```python
"""
XOR Gate using perceptron
(A OR B) AND (NOT(A AND B))
"""
import numpy as np


class Or_perceptron:
    def __init__(self):
        self.weights = np.array([0.5, 0.5])
        self.threshold = 0

    def execute(self, inputs):
        self.summation(inputs)
        return self.activation()

    def summation(self, inputs):
        self.inputs = inputs
        self.sum = np.dot(self.weights, self.inputs)

    def activation(self):
        if self.sum > self.threshold:
            return 1
        else:
            return 0


class And_perceptron:
    def __init__(self):
        self.weights = np.array([0.5, 0.6])
        self.threshold = 1

    def execute(self, inputs):
        self.summation(inputs)
        return self.activation()

    def summation(self, inputs):
        self.inputs = inputs
        self.sum = np.dot(self.weights, self.inputs)

    def activation(self):
        if self.sum > self.threshold:
            return 1
        else:
            return 0


class Not_preceptron:
    def __init__(self):
        self.weights = -0.5
        self.threshold = 0

    def execute(self, input):
        self.summation(input)
        return self.activation()

    def summation(self, input):
        self.input = input
        self.sum = self.weights * self.input

    def activation(self):
        if self.sum >= self.threshold:
            return 1
        else:
            return 0


class Xor_perceptron:
    def __init__(self, input):
        self.input = input
        self.or_perceptron()
        self.nand_perceptron()
        self.and_perceptron()

    def or_perceptron(self):
        or_res = Or_perceptron()
        self.or_res = or_res.execute(self.input)

    def nand_perceptron(self):
        and_res = And_perceptron()
        not_res = Not_preceptron()
        self.nand_res = not_res.execute(and_res.execute(self.input))

    def and_perceptron(self):
        and_res = And_perceptron()
        self.res = and_res.execute(np.array([self.or_res, self.nand_res]))

    def __str__(self):
        return f"Output Corresponding to input {self.input} is {self.res}"

if __name__ == '__main__':
    input1 = np.array([0, 0])
    input2 = np.array([0, 1])
    input3 = np.array([1, 0])
    input4 = np.array([1, 1])

    inputs=[input1,input2,input3,input4]
    for item in inputs:
        xor = Xor_perceptron(item)
        print(xor)
```
